```python
import pyclesperanto_prototype as cle
import stackview

cle.select_device("RTX")
```

```python
image = cle.imread("../../data/membranes.tif")
image
```

```python
def aqb(image, radius:float=1):
    #perc_image = 
    return cle.above_quantile_box(image, radius_x=radius, radius_y=radius)
    #cle.asarray(perc_image)

stackview.interact(aqb, image)
```

```python
cle.standard_deviation_box(image, radius_x=2.0, radius_y=2.0)
```

```python
def std(image, radius:int=1):
    return cle.standard_deviation_box(image, radius_x=radius, radius_y=radius)

stackview.interact(std, image)
```

```python
def sog(image, sigma):
    blurred = cle.gaussian_blur(image, sigma_x=sigma, sigma_y=sigma)
    return cle.sobel(blurred)

stackview.interact(sog, image)
```
